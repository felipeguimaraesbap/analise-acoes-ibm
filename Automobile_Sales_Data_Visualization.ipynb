```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import requests
import io
%matplotlib inline
```

## Data Loading and Preparation

```python
# Download do dataset adaptado para ambiente local
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/d51iMGfp_t0QpO30Lym-dw/automobile-sales.csv"
response = requests.get(URL)
response.raise_for_status()
df = pd.read_csv(io.StringIO(response.text))
print('Data downloaded and read into a dataframe!')
display(df.head())
```

### TASK 1.1: Annual Sales Fluctuations Line Chart

```python
# TASK 1.1: Line chart showing Average automobile sales fluctuate from year to year
df_line = df.groupby('Year')['Automobile_Sales'].mean()

plt.figure(figsize=(12, 6))
df_line.plot(kind='line', marker='o')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')

# Configurando os ticks com todos os anos para melhor identificação
plt.xticks(df['Year'].unique(), rotation=45)

# Anotações para dois anos de recessão marcantes (ex: 1980 e 2008)
plt.text(1980, df_line[1980] + 100, "1980 Recession", color='red', weight='bold')
plt.text(2008, df_line[2008] + 100, "2008 Recession", color='red', weight='bold')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### TASK 1.2: Advertising Expenditure Trends vs Sales Correlation

```python
# TASK 1.2: Scatter plot analyzing Advertising Expenditure vs Automobile Sales during Non-Recession Period
df_non_recession = df[df['Recession'] == 0]

plt.figure(figsize=(10, 6))
plt.scatter(df_non_recession['Advertising_Expenditure'], df_non_recession['Automobile_Sales'], alpha=0.7, color='green')
plt.xlabel('Advertising Expenditure (USD)')
plt.ylabel('Automobile Sales')
plt.title('Relationship between Advertising Expenditure and Sales during Non-Recession Periods')
plt.grid(True)
plt.show()

# Insight text block:
# Durante períodos de normalidade econômica, existe uma correlação fortemente positiva entre investimentos em marketing
# e o volume de vendas de automóveis, mostrando o impacto direto das campanhas publicitárias no comportamento do consumidor.
```

### TASK 1.3: Sales Trend Comparison Per Vehicle Type (Seaborn)

```python
# TASK 1.3: Compare sales trend per vehicle type using Seaborn (Recession vs Non-Recession)
plt.figure(figsize=(12, 6))
sns.barplot(x='Vehicle_Type', y='Automobile_Sales', hue='Recession', data=df)
plt.xlabel('Vehicle Type')
plt.ylabel('Average Automobile Sales')
plt.title('Vehicle Sales Comparison: Recession (1) vs Non-Recession (0)')
plt.show()

# Insight text block:
# As vendas despencam drasticamente de forma geral durante recessões. No entanto, carros econômicos e de uso familiar intermediário
# (como Mediumfamilycar e Smallfamilycar) sustentam volumes maiores proporcionalmente se comparados a modelos esportivos ou de luxo.
```

### TASK 1.4: Subplotting GDP Variations

```python
# TASK 1.4: Subplotting variations in GDP during recession and non-recession periods
df_rec = df[df['Recession'] == 1]
df_non_rec = df[df['Recession'] == 0]

fig = plt.figure(figsize=(14, 6))

# Subplot 1: Recession
ax0 = fig.add_subplot(1, 2, 1)
sns.lineplot(x='Year', y='GDP', data=df_rec, ax=ax0, marker='o', color='red')
ax0.set_title('GDP Variation During Recession Periods')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')

# Subplot 2: Non-Recession
ax1 = fig.add_subplot(1, 2, 2)
sns.lineplot(x='Year', y='GDP', data=df_non_rec, ax=ax1, marker='o', color='blue')
ax1.set_title('GDP Variation During Non-Recession Periods')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')

plt.tight_layout()
plt.show()
```

### TASK 1.5: Seasonality Impact on Automobile Sales (Bubble Plot)

```python
# TASK 1.5: Bubble plot displaying seasonality impact on Automobile Sales for Non-Recession Years
df_non_recession = df[df['Recession'] == 0]

plt.figure(figsize=(12, 7))
sns.scatterplot(x='Month', y='Automobile_Sales', size='Seasonality_Weight', 
                sizes=(20, 400), data=df_non_recession, hue='Seasonality_Weight', palette='viridis', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Seasonality impact on Automobile Sales')
plt.grid(True)
plt.show()
```

### TASK 1.6: Consumer Confidence & Prices vs Sales Scatter Plots

```python
# TASK 1.6: Scatter plots for Consumer Confidence vs Sales AND Vehicle Price vs Sales during Recessions
df_rec = df[df['Recession'] == 1]

# Gráfico 1: Confiança do Consumidor
plt.figure(figsize=(10, 5))
plt.scatter(df_rec['Consumer_Confidence'], df_rec['Automobile_Sales'], color='purple', alpha=0.7)
plt.xlabel('Consumer Confidence')
plt.ylabel('Automobile Sales')
plt.title('Consumer Confidence and Automobile Sales during Recessions')
plt.grid(True)
plt.show()

# Gráfico 2: Preço do Veículo
plt.figure(figsize=(10, 5))
plt.scatter(df_rec['Price'], df_rec['Automobile_Sales'], color='darkorange', alpha=0.7)
plt.xlabel('Vehicle Price')
plt.ylabel('Automobile Sales')
plt.title('Relationship between Vehicle Price and Sales during Recessions')
plt.grid(True)
plt.show()
```

### TASK 1.7: Advertising Expenditure Share Pie Chart

```python
# TASK 1.7: Pie chart displaying total advertising expenditure portion (Recession vs Non-Recession)
exp_data = df.groupby('Recession')['Advertising_Expenditure'].sum()

plt.figure(figsize=(8, 8))
plt.pie(exp_data, labels=['Non-Recession', 'Recession'], autopct='%1.1f%%', startangle=90, colors=['skyblue', 'salmon'])
plt.title('XYZAutomotives Advertising Expenditure Share')
plt.show()

# Insight text block:
# A empresa investe massivamente mais em marketing em tempos de estabilidade econômica (90%+). Durante crises,
# os orçamentos de publicidade sofrem severos cortes preventivos de custos.
```

### TASK 1.8: Advertisement Expenditure Share per Vehicle Type

```python
# TASK 1.8: Pie chart displaying total Advertisement expenditure for each vehicle type during Recession
df_rec = df[df['Recession'] == 1]
veh_exp = df_rec.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

plt.figure(figsize=(8, 8))
plt.pie(veh_exp, labels=veh_exp.index, autopct='%1.1f%%', startangle=140)
plt.title('Advertisement Expenditure Share per Vehicle Type During Recessions')
plt.show()
```

### TASK 1.9: Unemployment Rate Effect on Vehicle Type and Sales

```python
# TASK 1.9: Lineplot analyzing the effect of unemployment rate on vehicle type and sales during Recession
# Criando uma cópia explícita para evitar avisos de SettingWithCopy do Pandas
df_rec = df[df['Recession'] == 1].copy()

# Remove espaços em branco invisíveis que possam existir nos nomes das colunas
df_rec.columns = df_rec.columns.str.strip()

# Encontra a coluna de desemprego dinamicamente na lista (contendo 'unemployment')
unemp_col = [c for c in df_rec.columns if 'unemployment' in c.lower()][0]

plt.figure(figsize=(12, 6))

# USANDO A VARIÁVEL CORRETA (unemp_col) NO EIXO X
sns.lineplot(x=unemp_col, y='Automobile_Sales', hue='Vehicle_Type', data=df_rec, marker='o', errorbar=None)

plt.xlabel('Unemployment Rate')
plt.ylabel('Automobile Sales')
plt.title('Effect of Unemployment Rate on Vehicle Type and Sales')
plt.grid(True)
plt.show()
```
